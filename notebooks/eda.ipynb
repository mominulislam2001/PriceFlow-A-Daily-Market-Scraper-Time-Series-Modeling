```python
import matplotlib.pyplot as plt
```

```python
# Plot retail prices
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
plt.plot(df.index, df['min_retail_price'], label='Min Retail Price')
plt.plot(df.index, df['max_retail_price'], label='Max Retail Price')
plt.plot(df.index, df['avg_retail_price'], label='Avg Retail Price')
plt.xlabel('Date')
plt.ylabel('Retail Price (Kilogram)')
plt.title('Retail Prices Over Time')
plt.legend()

# Plot wholesale prices
plt.subplot(2, 1, 2)
plt.plot(df.index, df['min_wholesale_price'], label='Min Wholesale Price')
plt.plot(df.index, df['max_wholesale_price'], label='Max Wholesale Price')
plt.plot(df.index, df['avg_wholesale_price'], label='Avg Wholesale Price')
plt.xlabel('Date')
plt.ylabel('Wholesale Price (Quintal)')
plt.title('Wholesale Prices Over Time')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
# Create traces for retail prices
retail_trace_min = go.Scatter(x=df.index, y=df['min_retail_price'], mode='lines+markers', name='Min Retail Price')
retail_trace_max = go.Scatter(x=df.index, y=df['max_retail_price'], mode='lines+markers', name='Max Retail Price')
retail_trace_avg = go.Scatter(x=df.index, y=df['avg_retail_price'], mode='lines+markers', name='Avg Retail Price')


# Create subplots
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=('Retail Prices Over Time', 'Wholesale Prices Over Time'))

# Add traces to the subplots for retail prices
fig.add_trace(retail_trace_min, row=1, col=1)
fig.add_trace(retail_trace_max, row=1, col=1)
fig.add_trace(retail_trace_avg, row=1, col=1)


# Show plot
fig.show()
```
